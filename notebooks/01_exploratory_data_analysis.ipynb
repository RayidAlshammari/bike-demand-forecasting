```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
df = pd.read_csv('../data/raw/SeoulBikeData.csv', encoding='unicode_escape')
```

```python
display(df.head())
```

```python
display(df.info())
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Rented Bike Count'], bins=50, kde=True, color='blue')
plt.title('Distribution of Rented Bike Count (Target Variable)')
plt.xlabel('Rented Bike Count')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Hour', y='Rented Bike Count', data=df)
plt.title('Bike Demand by Hour of the Day (Identifying Peaks)')
plt.xlabel('Hour of the Day')
plt.ylabel('Rented Bike Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Temperature(°C)', y='Rented Bike Count', data=df, alpha=0.5)
plt.title('Temperature vs Bike Demand')
plt.xlabel('Temperature (°C)')
plt.ylabel('Rented Bike Count')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```
